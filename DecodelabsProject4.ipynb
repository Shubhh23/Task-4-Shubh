```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("Dataset for Data Analytics.xlsx")
df.head()
```

```python
df['Product'].value_counts().plot(kind='bar')

plt.title("Product Demand Analysis")
plt.xlabel("Product")
plt.ylabel("Number of Orders")

plt.show()
```

Business Question:
Which product is most preferred by customers?

Observation:
Printer has the highest number of orders (181), while Phone has the lowest number of orders (156).

Insight:
Customers show the highest demand for Printers.

Recommendation:
Maintain sufficient Printer inventory and focus marketing efforts on high-demand products.

```python
df['PaymentMethod'].value_counts().plot(kind='bar')

plt.title("Payment Method Preference")
plt.xlabel("Payment Method")
plt.ylabel("Number of Orders")

plt.show()
```

Business Question:
Which payment method is preferred by customers?

Observation:
Online payment has the highest number of transactions.

Insight:
Customers prefer digital payment methods over traditional payment options.

Recommendation:
Enhance the online payment experience and offer digital payment incentives.

```python
df['OrderStatus'].value_counts().plot(kind='bar')

plt.title("Order Status Distribution")
plt.xlabel("Order Status")
plt.ylabel("Number of Orders")

plt.show()
```

Business Question:
What is the distribution of order statuses?

Observation:
Cancelled orders slightly exceed Delivered orders.

Insight:
A considerable number of orders are not reaching successful completion.

Recommendation:
Investigate cancellation causes and improve order fulfillment processes.

```python
df['ReferralSource'].value_counts().plot(kind='bar')

plt.title("Customer Acquisition Sources")
plt.xlabel("Referral Source")
plt.ylabel("Number of Customers")

plt.show()
```

Business Question:
Which marketing channel attracts the most customers?

Observation:
Instagram contributes the highest number of customer referrals.

Insight:
Instagram is the most effective customer acquisition channel.

Recommendation:
Increase investment in Instagram marketing campaigns to attract more customers.

```python
df['TotalPrice'].hist(bins=10)

plt.title("Distribution of Order Values")
plt.xlabel("Total Price")
plt.ylabel("Frequency")

plt.show()
```

Business Question:
How are customer order values distributed?

Observation:
Most orders fall within lower price ranges, while a few orders have very high values.

Insight:
The business generates revenue primarily through many low-value orders and a small number of high-value purchases.

Recommendation:
Implement upselling and cross-selling strategies to increase average order value.

Final Business Insights

1. Printer is the most demanded product among customers.

2. Online payment is the preferred payment method.

3. Instagram is the strongest customer acquisition channel.

4. Order cancellations require further investigation to improve customer satisfaction.

5. Most customers place lower-value orders, while a few high-value orders contribute significantly to revenue.

6. Data visualization helped transform raw data into actionable business insights.
